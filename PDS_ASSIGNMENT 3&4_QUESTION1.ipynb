```python
#Question1 a)
import requests
import pandas as pd

def fetch_random_users():
    """Fetches random user data from the Random User Generator API."""
    url = "https://randomuser.me/api/?results=5"  # Fetch 5 random users
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print("Failed to retrieve data:", response.status_code)
        return None

def parse_users(data):
    """Parses JSON data to extract user details and returns a pandas DataFrame."""
    user_list = data['results']
    users = []
    for user in user_list:
        name = f"{user['name']['title']} {user['name']['first']} {user['name']['last']}"
        email = user['email']
        country = user['location']['country']
        users.append({'Name': name, 'Email': email, 'Country': country})
    return pd.DataFrame(users)

def display_dashboard(users_df):
    """Displays user data as a DataFrame."""
    print("\nUser Dashboard:")
    display(users_df)  # Using Jupyter's display function for better formatting

def main():
    data = fetch_random_users()
    if data:
        users_df = parse_users(data)
        display_dashboard(users_df)
    else:
        print("No data to display.")

# Run the main function
main()
```

```python
#QUESTION1 b
from jupyter_dash import JupyterDash
import dash
from dash import html, dcc
from dash.dependencies import Input, Output
import pandas as pd
import requests
import plotly.graph_objs as go

# Define the function to fetch random user data
def fetch_random_users():
    response = requests.get("https://randomuser.me/api/?results=5")
    if response.status_code == 200:
        return response.json()['results']
    else:
        return []

# Function to parse user data
def parse_users(users):
    return pd.DataFrame([{
        "Title": f"{user['name']['title']}",
        "First Name": f"{user['name']['first']}",
        "Last Name": f"{user['name']['last']}",
        "Email": user['email'],
        "Country": user['location']['country']
    } for user in users])

# Instantiate the Dash app
app = JupyterDash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Random User Dashboard"),
    html.Button("Refresh Data", id='refresh-data', n_clicks=0),
    dcc.Graph(id='user-table')
])

# Define the callback to update the graph
@app.callback(
    Output('user-table', 'figure'),
    Input('refresh-data', 'n_clicks')
)
def update_output(n_clicks):
    if n_clicks > 0:
        users = fetch_random_users()
        df = parse_users(users)
        figure = go.Figure(data=[go.Table(
            header=dict(values=list(df.columns),
                        fill_color='paleturquoise',
                        align='left'),
            cells=dict(values=[df[col] for col in df.columns],
                       fill_color='lavender',
                       align='left'))
        ])
        return figure
    return go.Figure()  # Return empty figure for the initial load

# Run the app inside the notebook
app.run_server(mode='inline')
```
